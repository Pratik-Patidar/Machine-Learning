```python
import numpy as np
a=np.array([[[1,2,3],[4,5,6]],[[1,2,3],[4,5,6]],[[1,2,3],[4,5,6]]])
print(a)
print(a.shape)
print(a.ndim)
print(a.size)
```

```python

a=np.array([[[1,2,3],[4,5,6j]],[[1,2,3],[4,5,6]],[[1,2,3],[4,5,6]]],dtype="complex")
print(a)
print(a.shape)
print(a.ndim)
print(a.size)
print(a.dtype)
```
